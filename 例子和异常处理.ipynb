## 1.爬取网页图片
下面的例子展示了re.findall的一种用法，可以直接返回括号中数据的列表。

这个程序可以打印一个网站上的图片，我是想打印点击进去后的高清图，有时间修改一下。

```python
# -*- coding: utf-8 -*-
#

import urllib.request
import os
import time
import re


def url_open(url):
    req = urllib.request.Request(url)
    req.add_header('User-Agent',
                   'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36')

    response = urllib.request.urlopen(req)
    html = response.read()

    # print(html)
    return html


def find_imgs(url):
    html = url_open(url).decode('utf-8')
    p = r'<img src="([^"]+.jpg)"'
    img_list = re.findall(p, html)

    # 加上http头
    for i in range(len(img_list)):
        img_list[i] = 'http:'+img_list[i]

    # print(img_list)

    return img_list

def save_img(folder, img_addrs):
    for each in img_addrs:
        filename = each.split('/')[-1]
        with open(filename, 'wb') as f:
            img = url_open(each)
            f.write(img)
    return

if __name__ == '__main__':
    folder = 'Beau'
    if not os.path.exists(folder):
        os.mkdir(folder)
    os.chdir(folder)
    url = 'https://www.ivsky.com/tupian/haijing_v49385/'
    img_list = find_imgs(url)
    save_img('bea', img_list)
    print("Completed!\n")
```

## 2.爬取ip地址

注意这里:
~~~py
p = r'(?:(?:[0,1]?\d?\d|2[0-4]\d|25[0-5])\.){3}(?:\b(?:[0,1]?\d?\d|2[0-4]\d|25[0-5])\b)' 

# 这里不能用下面。。。
p1= r'(?:(?:[0,1]?\d?\d|2[0-4]\d|25[0-5])\.){3}(?:(?:[0,1]?\d?\d|2[0-4]\d|25[0-5])$)'

需要
$ : 匹配字符串的结束位置，但是这里不能用。。。
或
\b : 匹配一个单词边界，单词被定义为Unicode的字母数字或下横线字符
否则：
n[12]: re.search(p1, '199.138.2.201')
Out[12]: <_sre.SRE_Match object; span=(0, 12), match='199.138.2.20'>
~~~

```python
# -*- coding: utf-8 -*-
#

import urllib.request
import os
import time
import re


def url_open(url):
    req = urllib.request.Request(url)
    req.add_header('User-Agent',
                   'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36')

    response = urllib.request.urlopen(req)
    html = response.read()

    # print(html)
    return html


def find_ips(url):
    html = url_open(url).decode('utf-8')

    # 这里加上?:表示不捕获元组
    p = r'(?:(?:[0,1]?\d?\d|2[0-4]\d|25[0-5])\.){3}(?:\b(?:[0,1]?\d?\d|2[0-4]\d|25[0-5])\b)'
    ip_list = re.findall(p, html)

    # print(ip_list)

    return ip_list


if __name__ == '__main__':
    folder = 'Beau'
    if not os.path.exists(folder):
        os.mkdir(folder)
    os.chdir(folder)
    url = 'https://www.xicidaili.com/'
    ip_list = find_ips(url)
    
    print(ip_list)
```

# 异常处理

域名解析失败：URLError

HTTPError

Http Status codes: https://www.restapitutorial.com/httpstatuscodes.html

## 1.处理异常的第一种写法

```python
from urllib.request import Request, urlopen
from urllib.error import URLError, HTTPError
req = Request("https://www.taobao.com/")

try:
    response = urlopen(req)
except HTTPError as e:
    # HTTPError is a subset of URLError, it must be run at first
    print('The server cound\'t fulfill the request.')
    print('Error code: ',e.code)
except URLError as e:
    print('We failed to reach a server. ')
    print('Reason: ', e.reason)
else:
    # You can use the else keyword to define a block of code to be executed if no errors were raised
    html = response.read().decode('utf-8')
    print(html)
```

## 2.第二种写法(推荐)

```python
from urllib.request import Request, urlopen
from urllib.error import URLError, HTTPError
req = Request("https://www.taoba1o.com/")

try:
    response = urlopen(req)
except URLError as e:
    if hasattr(e, 'reason'):  # hasattr是否有该属性
        print('We failed to reach a server. ')
        print('Reason: ', e.reason)
    elif hasattr(e, 'code'):
        print('The server cound\'t fulfill the request.')
        print('Error code: ',e.code)
else:
    # You can use the else keyword to define a block of code to be executed if no errors were raised
    html = response.read().decode('utf-8')
    print(html)
# everything is fine
```
