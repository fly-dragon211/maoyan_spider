```python
import requests
from lxml import etree

res = requests.get("https://www.pearvideo.com/category_8")
print(res.text)
```

## 把网页变成xPath结构

```python
res_xpath = etree.HTML(res.text)
```

### /text提取文字

```python
res_xpath.xpath('/html/body/div/text()') # 绝对路径
```

### //提取任意子节点

```python
res_xpath.xpath('//title/text()')
```

```python
res_xpath.xpath('//a[@class="menu" and @href="category_10"]/text()') 
# 多个属性需要加关键字and, xpath支持运算符
```

### 提取属性

注意有些属性是渲染出来的，不能用可以换一个属性

使用中括号[]指定提取第几个

'*' 代表所有元素

```python
res_xpath.xpath('//*[@id="listvideoListUl"]/li[@class="categoryem "]')
```

二次提取，注意需要加'.'表示相对路径，否则就会从头开始找。

```python
for li in res_xpath.xpath('//*[@id="listvideoListUl"]/li[@class="categoryem "]'):
    print('https://www.pearvideo.com/' +
          li.xpath('.//a[@class="vervideo-lilink actplay"]/@href')[0]) 
```

### 使用正则提取数据

```python
res = requests.get("https://www.pearvideo.com/video_1611397")
```

这里查看网页源代码发现，视频链接并不在xpath可以识别的范围，所以使用正则提取。

```python
import re

re.findall?
```

```python
# re.S忽略换行符
# . 任意一个字符
# * 1或多歌字符
# ? 非贪婪匹配

video_url = re.findall(r'srcUrl="(.*?)"', res.text, re.S)
print(video_url)
```

```python
# 拿到的是二进制代码。
# 可见即可爬

video_res = requests.get(video_url[0])
```

```python
# 保存到文件
with open("a.mp4", mode="wb") as f:
    # response.content
    f.write(video_res.content)
```
